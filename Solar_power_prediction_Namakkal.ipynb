```python
import warnings
import pandas as pd
import MySQLdb
import datetime
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import time
import requests

# MySQL connection setup
db = MySQLdb.connect(host="localhost", port=3306, user="root", db="prediction11", unix_socket="/Applications/XAMPP/xamppfiles/var/mysql/mysql.sock")
cur = db.cursor()

def fetch_recent_weather(latitude, longitude):
    """
    Fetch last 1 day hourly weather data from Open-Meteo API for given location.
    Returns a DataFrame with features needed for training.
    """
    end_date = datetime.datetime.utcnow().date()
    start_date = end_date - datetime.timedelta(days=1)

    url = f"https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": start_date.isoformat(),
        "end_date": end_date.isoformat(),
        "hourly": "temperature_2m,relative_humidity_2m,cloudcover,windspeed_10m",
        "timezone": "Asia/Kolkata"
    }

    response = requests.get(url, params=params)
    data = response.json()
    df = pd.DataFrame(data['hourly'])

    # Convert time to datetime and create time features
    df['time'] = pd.to_datetime(df['time'])
    df['year'] = df['time'].dt.year
    df['month'] = df['time'].dt.month
    df['day'] = df['time'].dt.day
    df['hour'] = df['time'].dt.hour
    df['minute'] = df['time'].dt.minute

    # Estimate GHI from cloud cover (simplistic)
    df['GHI'] = (1 - df['cloudcover'] / 100) * 1000

    # Select and rename columns for consistency with training data
    df_model = df[['year', 'month', 'day', 'hour', 'minute', 'temperature_2m', 'GHI', 'relative_humidity_2m']].copy()
    df_model.rename(columns={'temperature_2m':'Temperature', 'relative_humidity_2m': 'RelativeHumidity'}, inplace=True)

    # Save recent actual data to CSV
    df_model.to_csv('recent_actual_data.csv', index=False)

    return df_model

def listToString(s):
    str1 = ""
    for ele in s:
        str1 += ele
    return str1

latitude = 11.2189   # Namakkal
longitude = 78.1670

while True:
    try:
        # Suppress Warnings
        def fxn():
            warnings.warn("deprecated", DeprecationWarning)
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            fxn()

        # Fetch recent actual data and save
        fetch_recent_weather(latitude, longitude)

        # Load historical and recent data
        df_hist = pd.read_csv("merged_dataset.csv")
        df_hist = df_hist.fillna(0)
        df_recent = pd.read_csv("recent_actual_data.csv")

        # Combine datasets for training
        df_combined = pd.concat([df_hist, df_recent], ignore_index=True)

        # Features and targets
        feature_cols = ['year', 'month', 'day', 'hour', 'minute']
        target_temp = 'Temperature'
        target_ghi = 'GHI'

        X = df_combined[feature_cols].values
        Y_temp = df_combined[target_temp].values
        Y_ghi = df_combined[target_ghi].values

        # Train-test split
        x_train, x_test, y_temp_train, y_temp_test = train_test_split(X, Y_temp, random_state=42)
        x_train, x_test, y_ghi_train, y_ghi_test = train_test_split(X, Y_ghi, random_state=42)

        # Train models
        rfc1 = RandomForestRegressor()
        rfc2 = RandomForestRegressor()
        rfc1.fit(x_train, y_temp_train)
        rfc2.fit(x_train, y_ghi_train)

        # Predict for 15 min ahead
        time_15min_later = datetime.datetime.now() + datetime.timedelta(minutes=15)
        time_15min_later_str = time_15min_later.strftime("%Y-%m-%d %H:%M")
        now = [time_15min_later.year, time_15min_later.month, time_15min_later.day, time_15min_later.hour, time_15min_later.minute]

        # Make predictions
        temp = rfc1.predict([now])[0]
        ghi = rfc2.predict([now])[0]

        # Power calculation
        f = 0.18 * 7.4322 * ghi
        insi = temp - 25
        midd = 0.95 * insi
        power = f * midd

        # Prepare SQL data and insert
        data = (time_15min_later_str, float(temp), float(ghi), float(power))
        sql = """INSERT INTO Solar_power_prediction (time_updated, Temperature, GHI, power) VALUES (%s, %s, %s, %s)"""
        cur.execute(sql, data)
        db.commit()
        print(f"Inserted prediction for {time_15min_later_str}")

    except Exception as e:
        print(f"Error occurred: {str(e)}")
        db.rollback()

    time.sleep(60)  # Sleep for 60 seconds before next iteration

# Close connection (usually not reached because loop is infinite)
cur.close()
db.close()
```
